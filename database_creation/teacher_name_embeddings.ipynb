```python
import langchain
from langchain_community.document_loaders import DataFrameLoader
import json
import pandas as pd
import getpass
import os
from pinecone import Pinecone, ServerlessSpec
from langchain_pinecone import PineconeVectorStore
from langchain_openai import OpenAIEmbeddings 
import openai 
from langchain_core.documents import Document
```

```python
df = pd.read_csv('../data/df_lectures.csv')
```

```python
#Setting up keys
pinecone_api_key = os.environ.get("pinecone_API")

pc = Pinecone(api_key=pinecone_api_key)

api_key = os.getenv("OPENAI_API_KEY")
openai.api_key = api_key
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

pc.create_index(
    name= 'teacher-names',
    dimension=1536,
    metric="cosine",
    spec=ServerlessSpec(cloud="aws", region="us-east-1"),
)

```

```python
df.dropna(subset=['Instructor'], inplace=True)
instructor_list = df['Instructor'].unique()
print(len(instructor_list))
```

```python
documents = [Document(page_content = doc) for doc in instructor_list]
```

```python
index = pc.Index("teacher-names")
vector_store = PineconeVectorStore(index , embedding = embeddings)
```

```python
vector_store.delete(delete_all = True)
```

```python

vector_store.add_documents(documents)

```
